```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('stu_perf_clean.csv')
```

```python
data
```

```python
plt.figure(figsize=(10, 6))
sns.ecdfplot(data['math score'], marker='o', linestyle='-', color='orange')
plt.title('Cumulative Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Cumulative Probability')
plt.show()
```

```python
plt.figure(figsize=(10, 6))


sns.jointplot(x='reading score', y='writing score', data=data, kind='hex', color='green')
plt.suptitle('Relationship between Reading and Writing Scores', y=1.02)
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='reading score', y='math score', data=data, color='blue')
plt.title('Reading Score vs Math Score')
plt.xlabel('Reading Score')
plt.ylabel('Math Score')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='parental level of education', y='math score', hue='race/ethnicity', data=data, palette='viridis')
plt.title('Math Scores by Parental Level of Education and Ethnicity')
plt.xlabel('Parental Level of Education')
plt.ylabel('Math Score')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Ethnicity', loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='parental level of education', hue='test preparation course', data=data, palette='coolwarm')
plt.title('Test Preparation Course by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Test Preparation Course', loc='upper right')
plt.show()
```
