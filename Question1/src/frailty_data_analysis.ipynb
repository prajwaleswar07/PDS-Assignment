```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('clean_data_Fraility.csv')
```

```python
data
```

```python
plt.pie(data['Age'].value_counts(), labels=data['Age'].unique(), autopct='%1.0f%%', startangle=90)
plt.title('Distribution of Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Frailty', y='Age', data=data, inner='quartile', palette='pastel')
plt.title('Age Distribution by Frailty')
plt.xlabel('Frailty')
plt.ylabel('Age')
plt.show()




```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(data['Age'], data['Grip strength'], marker='o', linestyle='-', label='Grip Strength', color='blue')
ax1.set_xlabel('Age')
ax1.set_ylabel('Grip Strength', color='blue')
ax1.tick_params('y', colors='blue')

ax2 = ax1.twinx()
ax2.plot(data['Age'], data['Height (Inches)'], marker='s', linestyle='-', label='Height', color='green')
ax2.set_ylabel('Height (Inches)', color='green')
ax2.tick_params('y', colors='green')

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

plt.title('Grip Strength and Height by Age')
plt.show()
```
